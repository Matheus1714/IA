```python
(size, _) = df_domains.shape

domains = []
labels_domains = []

paths = []
labels_paths = []


for i in tqdm(range(size)):
    
    domain = literal_eval(df_domains.iloc[i].tolist()[0])
    label_domain = df_domains.iloc[i].tolist()[1]
    
    path = literal_eval(df_paths.iloc[i].tolist()[0])
    label_path = df_paths.iloc[i].tolist()[1]
    
    domains.append(domain)
    labels_domains.append(label_domain)
    
    paths.append(path)
    labels_paths.append(label_path)
```

```python
(size, _) = df_domains_related.shape

domains_related_queries = []
labels_domains_related = []

for i in tqdm(range(size)):
    
    domain = literal_eval(df_domains_related.iloc[i].tolist()[0])
    label_domain = df_domains_related.iloc[i].tolist()[1]
    
    domains_related_queries.append(domain)
    labels_domains_related.append(label_domain)
```

```python
(size, _) = df_paths_related.shape

paths_related_queries = []
labels_paths_related_queries = []

for i in tqdm(range(size)):
    
    path = literal_eval(df_paths_related.iloc[i].tolist()[0])
    label_path = df_paths_related.iloc[i].tolist()[1]
    
    paths_related_queries.append(path)
    labels_paths_related_queries.append(label_path)
```

```python
def jaccard_similarity(list1, list2):
    
    s1 = set(list1)
    s2 = set(list2)
    
    if len(s1.union(s2)) != 0:
        return len(s1.intersection(s2)) / len(s1.union(s2))
    
    else:
        return 0
```

```python
#Jaccard index between domains and related queries

size_domains_related_queries = len(domains_related_queries)

j_indexes_domain_related = []

labels_j_indexes_domain_related = labels_domains_related

for i in tqdm(range(size_domains_related_queries)):
    
    domain_related = domains_related_queries[i]
    
    registered_domain = domains[i]
    
    j_indexes_domain_related.append(jaccard_similarity(registered_domain, domain_related))
```

```python
# Jaccard index between paths and related queries

size_paths_related_queries = len(paths_related_queries)

j_indexes_paths_related = []

labels_j_indexes_path_related = labels_paths_related_queries

for i in tqdm(range(size_paths_related_queries)):
    
    path = paths[i]
    
    path_related = paths_related_queries[i]
    
    j_indexes_paths_related.append(jaccard_similarity(path_related, path))
```

```python
j_indexes_paths_related_to_domain = [] #Jaccard index related queries domain and paths

j_indexes_domain_related_to_paths = [] #Jaccard index related queries path and domain

j_indexes_domain_path = [] #Jaccard index between paths and domains

j_indexes_related_to_path_related_to_domain = []

size = len(paths_related_queries)

for i in tqdm(range(size)):
    
    j_indexes_paths_related_to_domain.append(jaccard_similarity(paths[i],domains_related_queries[i]))
    
    j_indexes_domain_related_to_paths.append(jaccard_similarity(paths_related_queries[i],domains[i]))
    
    j_indexes_domain_path.append(jaccard_similarity(paths[i],domains[i]))
    
    j_indexes_related_to_path_related_to_domain.append(jaccard_similarity(paths_related_queries[i],domains_related_queries[i]))
```

```python
from sklearn.preprocessing import label_binarize

j_indexes = []

j_indexes.extend(j_indexes_paths_related_to_domain)
j_indexes.extend(j_indexes_domain_related_to_paths)
j_indexes.extend(j_indexes_domain_path)
j_indexes.extend(j_indexes_related_to_path_related_to_domain)
j_indexes.extend(j_indexes_paths_related)
j_indexes.extend(j_indexes_domain_related)

labels = []

labels.extend(labels_paths)
labels.extend(labels_paths)
labels.extend(labels_paths)
labels.extend(labels_paths)
labels.extend(labels_paths)
labels.extend(labels_paths)

#              ,
#              j_indexes_domain_path,
#              j_indexes_related_to_path_related_to_domain,
#              j_indexes_paths_related,
#              j_indexes_domain_related,)

# j_indexes = [j_indexes_paths_related_to_domain,
#              j_indexes_domain_related_to_paths,
#              j_indexes_domain_path,
#              j_indexes_related_to_path_related_to_domain,
#              j_indexes_paths_related,
#              j_indexes_domain_related,
#             ]
# df_j_indexes = pandas.DataFrame(j_indexes)
# df_j_indexes
```

```python
from sklearn.model_selection import train_test_split

random_state = 42

train_features, test_features, train_labels, test_labels = train_test_split(
    j_indexes,
    labels,
    test_size = 0.25,
    random_state = random_state
)
```

```python
import numpy as np
from sklearn. ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, roc_auc_score

#Training model with random forest

n_estimators = 1000

rf = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')

rf.fit(np.array(train_features).reshape(-1,1), np.array(train_labels).reshape(-1,1))

#Testing model and finding error

predictions = rf.predict(np.array(test_features).reshape(-1,1))

probs = rf.predict_proba(np.array(test_features).reshape(-1,1))[:, 1]

matrix = confusion_matrix(np.array(test_labels).reshape(-1,1), predictions)

roc_value = roc_auc_score(np.array(test_labels).reshape(-1,1), probs)

print(matrix)
print(roc_value)
```
